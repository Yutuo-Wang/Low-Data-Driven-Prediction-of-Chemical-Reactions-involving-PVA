# PVA-ReAct Model Training Example

This example demonstrates how to train the PVA-ReAct model for chemical reaction prediction.

```python
# Import necessary libraries
import sys
import os
sys.path.append('../')

import pandas as pd
import torch
import matplotlib.pyplot as plt
from training.train_main import ReactionTrainer
from utils.data_loader import ReactionDataLoader
from utils.config import TrainingConfig
from features.dft_extractor import DFTFeatureExtractor
from transformers import AutoTokenizer
```

```python
# Configure training parameters
config = TrainingConfig(
    batch_size=16,
    hidden_dim=256,
    learning_rate=1e-4,
    epochs=100,
    early_stopping_patience=10,
    max_length=200
)

print("Training Configuration:")
print(f"- Batch Size: {config.batch_size}")
print(f"- Hidden Dimension: {config.hidden_dim}")
print(f"- Learning Rate: {config.learning_rate}")
print(f"- Epochs: {config.epochs}")
```

```python
# Initialize data loader
data_loader = ReactionDataLoader(
    data_path="../data/reaction_data.csv",
    test_size=0.1,
    val_size=0.1
)

# Load and preprocess data
train_data, val_data, test_data, scaler = data_loader.load_and_preprocess()

print(f"Data loading completed:")
print(f"- Training set: {len(train_data)} samples")
print(f"- Validation set: {len(val_data)} samples") 
print(f"- Test set: {len(test_data)} samples")
```

```python
# Initialize feature extractor and tokenizer
dft_extractor = DFTFeatureExtractor(basis='def2svp', xc='B3LYP')
tokenizer = AutoTokenizer.from_pretrained("seyonec/PubChem10M_SMILES_BPE_396_250")

# Create data loaders
train_loader, val_loader, test_loader = data_loader.create_dataloaders(
    train_data, val_data, test_data, tokenizer, dft_extractor, config.batch_size
)
```

```python
# Initialize trainer
trainer = ReactionTrainer(config, tokenizer, dft_extractor)

# Start training
print("Starting model training...")
training_history = trainer.train(train_loader, val_loader)

print("Training completed!")
```

```python
# Visualize training process
plt.figure(figsize=(15, 5))

# Training loss
plt.subplot(1, 3, 1)
plt.plot(training_history['train_loss'])
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)

# Validation R²
plt.subplot(1, 3, 2)
plt.plot(training_history['val_r2'])
plt.title('Validation R² Score')
plt.xlabel('Epoch')
plt.ylabel('R²')
plt.grid(True)

# Validation accuracy
plt.subplot(1, 3, 3)
plt.plot(training_history['val_accuracy'])
plt.title('Validation Accuracy (±5%)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Evaluate model on test set
test_metrics = trainer.evaluate(test_loader)

print("Test Set Evaluation Results:")
print(f"- R² Score: {test_metrics['r2']:.4f}")
print(f"- MSE: {test_metrics['mse']:.4f}")
print(f"- Accuracy (±5%): {test_metrics['accuracy']:.2%}")

# Save trained model
trainer.save_model('../models/pretrained_models/pva_react_model.pth')
print("Model saved!")
```
