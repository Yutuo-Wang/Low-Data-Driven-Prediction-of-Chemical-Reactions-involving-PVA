# PVA-ReAct Model Prediction Example

This example demonstrates how to use the trained PVA-ReAct model for chemical reaction prediction.

```python
# Import necessary libraries
import sys
import os
sys.path.append('../')

import torch
import pandas as pd
from inference.predict_main import ReactionPredictor
from features.dft_extractor import DFTFeatureExtractor
from transformers import AutoTokenizer
from rdkit import Chem
from rdkit.Chem import Draw
import matplotlib.pyplot as plt
```

```python
# Initialize predictor
model_path = "../models/pretrained_models/pva_react_model.pth"
predictor = ReactionPredictor(model_path)

print("Predictor initialized")
print(f"Loaded model: {model_path}")
```

## Example 1: Basic Reaction Prediction

```python
# Define example reactions
example_reactions = [
    {
        "name": "PVA-Styrene Grafting",
        "reactant": "CC(=C)C(=O)OC",  # Methyl methacrylate
        "product": "*C(CC(C)(C(=O)OC)*)C(O)*",  # Graft product
        "temperature": 80,
        "time": 4
    },
    {
        "name": "PVA Crosslinking", 
        "reactant": "B(O)O",  # Boric acid
        "product": "*B(O*)O*",  # Crosslinked product
        "temperature": 70,
        "time": 3
    }
]

# Make predictions
print("ðŸ”¬ Starting reaction prediction...")
for i, reaction in enumerate(example_reactions, 1):
    yield_pred, confidence = predictor.predict_yield(
        reaction["reactant"],
        reaction["product"], 
        reaction["temperature"],
        reaction["time"]
    )
    
    print(f"\nðŸ“Š Reaction {i}: {reaction['name']}")
    print(f"   - Reactant: {reaction['reactant']}")
    print(f"   - Product: {reaction['product']}")
    print(f"   - Predicted Yield: {yield_pred:.1f}%")
    print(f"   - Confidence: {confidence:.3f}")
    print(f"   - Conditions: {reaction['temperature']}Â°C, {reaction['time']}h")
```

## Example 2: Molecular Structure Visualization

```python
# Visualize molecular structures
def visualize_molecules(reactions):
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    
    for i, reaction in enumerate(reactions):
        # Reactant
        reactant_mol = Chem.MolFromSmiles(reaction["reactant"])
        if reactant_mol:
            axes[i, 0].imshow(Draw.MolToImage(reactant_mol, size=(300, 300)))
            axes[i, 0].set_title(f"Reactant: {reaction['name']}")
            axes[i, 0].axis('off')
        
        # Product
        product_mol = Chem.MolFromSmiles(reaction["product"])
        if product_mol:
            axes[i, 1].imshow(Draw.MolToImage(product_mol, size=(300, 300)))
            axes[i, 1].set_title(f"Product: {reaction['name']}")
            axes[i, 1].axis('off')
    
    plt.tight_layout()
    plt.show()

# Display molecular structures
visualize_molecules(example_reactions)
```

## Example 3: Batch Prediction

```python
# Batch prediction example
batch_reactions = [
    ("CCCO", "*C(CCO)*", 60, 2),  # Alcohol oxidation
    ("C=CC", "*C(CC)*", 120, 6),  # Olefin hydrogenation
    ("CCOC(=O)O", "*C(COC(=O)O)*", 90, 4),  # Esterification
]

print("Batch prediction results:")
results = []
for reactant, product, temp, time in batch_reactions:
    yield_pred, confidence = predictor.predict_yield(reactant, product, temp, time)
    results.append({
        'reactant': reactant,
        'product': product, 
        'temperature': temp,
        'time': time,
        'predicted_yield': yield_pred,
        'confidence': confidence
    })
    
    print(f"  - {reactant} â†’ {product}: {yield_pred:.1f}% (confidence: {confidence:.3f})")

# Convert to DataFrame for analysis
results_df = pd.DataFrame(results)
print(f"\nðŸ“Š Batch prediction statistics:")
print(f"   - Average yield: {results_df['predicted_yield'].mean():.1f}%")
print(f"   - Maximum yield: {results_df['predicted_yield'].max():.1f}%")
print(f"   - Minimum yield: {results_df['predicted_yield'].min():.1f}%")
```

## Example 4: DFT Feature Analysis

```python
# Analyze DFT features
dft_extractor = DFTFeatureExtractor()

print(" DFT Feature Analysis:")
for reaction in example_reactions:
    reactant_features = dft_extractor.calculate(reaction["reactant"])
    product_features = dft_extractor.calculate(reaction["product"])
    
    print(f"\n {reaction['name']}:")
    print(f"   Reactant - HOMO: {reactant_features[0]:.3f}, LUMO: {reactant_features[1]:.3f}")
    print(f"   Product  - HOMO: {product_features[0]:.3f}, LUMO: {product_features[1]:.3f}")
    
    # Calculate energy gap change
    reactant_gap = reactant_features[1] - reactant_features[0]
    product_gap = product_features[1] - product_features[0]
    gap_change = product_gap - reactant_gap
    
    print(f"   Energy gap change: {gap_change:.3f} eV")
    if gap_change > 0:
        print(f" Energy gap increased, reaction may be more stable")
    else:
        print(f"  Energy gap decreased, reaction activity may be enhanced")
```
