# PVA-ReAct Reaction Condition Optimization Example

This example demonstrates how to use the PVA-ReAct framework for inverse design and optimization of reaction conditions.

```python
# Import necessary libraries
import sys
import os
sys.path.append('../')

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from inference.condition_optimizer import ConditionOptimizer
from inference.predict_main import ReactionPredictor
from features.dft_extractor import DFTFeatureExtractor
from transformers import AutoTokenizer
import pandas as pd
```

```python
# Initialize optimizer
model_path = "../models/pretrained_models/pva_react_model.pth"
optimizer = ConditionOptimizer(model_path)

print("Condition optimizer initialized")
print("Starting reaction condition optimization...")
```

## Example 1: PVA Crosslinking Reaction Condition Optimization

```python
# Define target reaction
crosslinking_reaction = {
    "name": "PVA-Boric Acid Crosslinking",
    "reactant": "B(O)O",  # Boric acid
    "product": "*B(O*)O*",  # Crosslinked PVA
    "description": "Reversible crosslinking of PVA with boric acid for smart hydrogel preparation"
}

print(f"Optimization reaction: {crosslinking_reaction['name']}")
print(f"Description: {crosslinking_reaction['description']}")
print(f"Reactant: {crosslinking_reaction['reactant']}")
print(f"Product: {crosslinking_reaction['product']}")

# Optimize reaction conditions
optimal_temp, optimal_time, optimization_history = optimizer.optimize_conditions(
    crosslinking_reaction["reactant"],
    crosslinking_reaction["product"],
    init_temp=50,
    init_time=2,
    temp_bounds=(20, 120),
    time_bounds=(0.5, 24),
    max_iterations=50
)

print(f"\n Optimization completed!")
print(f" Optimal conditions:")
print(f"   - Temperature: {optimal_temp:.1f}Â°C")
print(f"   - Time: {optimal_time:.1f} hours")

# Predict yield under optimal conditions
optimal_yield, confidence = optimizer.predictor.predict_yield(
    crosslinking_reaction["reactant"],
    crosslinking_reaction["product"],
    optimal_temp,
    optimal_time
)

print(f"   - Predicted yield: {optimal_yield:.1f}%")
print(f"   - Confidence: {confidence:.3f}")
```

## Example 2: Optimization Process Visualization

```python
# Visualize optimization process
def plot_optimization_history(history):
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    # Yield optimization process
    axes[0, 0].plot(history['yield_history'], 'b-', linewidth=2)
    axes[0, 0].set_title('Yield Optimization Process', fontsize=14, fontweight='bold')
    axes[0, 0].set_xlabel('Iteration')
    axes[0, 0].set_ylabel('Predicted Yield (%)')
    axes[0, 0].grid(True, alpha=0.3)
    axes[0, 0].axhline(y=max(history['yield_history']), color='r', linestyle='--', 
                       label=f'Max Yield: {max(history["yield_history"]):.1f}%')
    axes[0, 0].legend()
    
    # Temperature evolution
    axes[0, 1].plot(history['temperature_history'], 'g-', linewidth=2)
    axes[0, 1].set_title('Temperature Optimization', fontsize=14, fontweight='bold')
    axes[0, 1].set_xlabel('Iteration')
    axes[0, 1].set_ylabel('Temperature (Â°C)')
    axes[0, 1].grid(True, alpha=0.3)
    
    # Time evolution
    axes[1, 0].plot(history['time_history'], 'm-', linewidth=2)
    axes[1, 0].set_title('Time Optimization', fontsize=14, fontweight='bold')
    axes[1, 0].set_xlabel('Iteration')
    axes[1, 0].set_ylabel('Time (hours)')
    axes[1, 0].grid(True, alpha=0.3)
    
    # Condition space exploration
    sc = axes[1, 1].scatter(history['temperature_history'], 
                           history['time_history'], 
                           c=history['yield_history'], 
                           cmap='viridis', s=50, alpha=0.7)
    axes[1, 1].set_title('Condition Space Exploration', fontsize=14, fontweight='bold')
    axes[1, 1].set_xlabel('Temperature (Â°C)')
    axes[1, 1].set_ylabel('Time (hours)')
    plt.colorbar(sc, ax=axes[1, 1], label='Yield (%)')
    
    # Mark optimal condition
    axes[1, 1].scatter(optimal_temp, optimal_time, color='red', s=200, 
                      marker='*', edgecolors='black', label='Optimal Condition')
    axes[1, 1].legend()
    
    plt.tight_layout()
    plt.show()

# Plot optimization history
plot_optimization_history(optimization_history)
```

## Example 3: Multi-Reaction Parallel Optimization

```python
# Define multiple reactions for optimization
multi_reactions = [
    {
        "name": "PVA-MMA Grafting",
        "reactant": "CC(=C)C(=O)OC",  # MMA
        "product": "*C(C(C)=C(C(=O)OC))*C(O)*",  # PVA-g-PMMA
        "init_temp": 70,
        "init_time": 3
    },
    {
        "name": "PVA-Acrylic Acid Hydrogel",
        "reactant": "C=CC(=O)O",  # Acrylic acid
        "product": "*C(CC(=O)O)*C(O)*",  # PVA/PAA hydrogel
        "init_temp": 60,
        "init_time": 4
    },
    {
        "name": "PVA Azidation",
        "reactant": "N=[N+]=[N-]",  # Azide group
        "product": "*C(C[N-])=[N+]=N",  # Azidated PVA
        "init_temp": 40,
        "init_time": 6
    }
]

print("Starting multi-reaction parallel optimization...")
optimization_results = []

for reaction in multi_reactions:
    print(f"\nðŸ”¬ Optimizing: {reaction['name']}")
    
    opt_temp, opt_time, _ = optimizer.optimize_conditions(
        reaction["reactant"],
        reaction["product"],
        init_temp=reaction["init_temp"],
        init_time=reaction["init_time"],
        max_iterations=30
    )
    
    # Predict yield
    pred_yield, confidence = optimizer.predictor.predict_yield(
        reaction["reactant"],
        reaction["product"],
        opt_temp,
        opt_time
    )
    
    optimization_results.append({
        'reaction': reaction['name'],
        'reactant': reaction['reactant'],
        'optimal_temperature': opt_temp,
        'optimal_time': opt_time,
        'predicted_yield': pred_yield,
        'confidence': confidence
    })
    
    print(f"Optimal conditions: {opt_temp:.1f}Â°C, {opt_time:.1f}h")
    print(f"Predicted yield: {pred_yield:.1f}% (confidence: {confidence:.3f})")

# Create results DataFrame
results_df = pd.DataFrame(optimization_results)
print(f"\nðŸ“ˆ Parallel optimization completed! Optimized {len(results_df)} reactions")
```

## Example 4: Optimization Results Analysis and Visualization

```python
# Analyze optimization results
print("ðŸ“Š Optimization Results Analysis:")
print(f"   - Average optimized yield: {results_df['predicted_yield'].mean():.1f}%")
print(f"   - Maximum yield: {results_df['predicted_yield'].max():.1f}% ({results_df.loc[results_df['predicted_yield'].idxmax(), 'reaction']})")
print(f"   - Minimum yield: {results_df['predicted_yield'].min():.1f}% ({results_df.loc[results_df['predicted_yield'].idxmin(), 'reaction']})")
print(f"   - Average temperature: {results_df['optimal_temperature'].mean():.1f}Â°C")
print(f"   - Average time: {results_df['optimal_time'].mean():.1f} hours")

# Visualize optimization results
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Yield distribution
axes[0, 0].bar(results_df['reaction'], results_df['predicted_yield'], 
               color=['skyblue', 'lightgreen', 'lightcoral'])
axes[0, 0].set_title('Optimized Yields by Reaction', fontsize=14, fontweight='bold')
axes[0, 0].set_ylabel('Predicted Yield (%)')
axes[0, 0].tick_params(axis='x', rotation=45)
for i, v in enumerate(results_df['predicted_yield']):
    axes[0, 0].text(i, v + 1, f'{v:.1f}%', ha='center', va='bottom', fontweight='bold')

# Condition distribution
x = np.arange(len(results_df))
width = 0.35
axes[0, 1].bar(x - width/2, results_df['optimal_temperature'], width, label='Temperature (Â°C)', color='orange')
axes[0, 1].bar(x + width/2, results_df['optimal_time'], width, label='Time (h)', color='purple')
axes[0, 1].set_title('Optimal Reaction Conditions', fontsize=14, fontweight='bold')
axes[0, 1].set_xticks(x)
axes[0, 1].set_xticklabels(results_df['reaction'])
axes[0, 1].legend()
axes[0, 1].tick_params(axis='x', rotation=45)

# Condition-yield relationship
scatter = axes[1, 0].scatter(results_df['optimal_temperature'], 
                            results_df['optimal_time'], 
                            c=results_df['predicted_yield'], 
                            s=200, cmap='viridis', alpha=0.7)
axes[1, 0].set_title('Condition-Yield Relationship', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Temperature (Â°C)')
axes[1, 0].set_ylabel('Time (hours)')
plt.colorbar(scatter, ax=axes[1, 0], label='Yield (%)')

# Add reaction labels
for i, row in results_df.iterrows():
    axes[1, 0].annotate(row['reaction'], 
                       (row['optimal_temperature'], row['optimal_time']),
                       xytext=(5, 5), textcoords='offset points',
                       fontsize=9, alpha=0.8)

# Confidence analysis
wedges, texts, autotexts = axes[1, 1].pie(results_df['confidence'], 
                                         labels=results_df['reaction'], 
                                         autopct='%1.2f', 
                                         colors=['gold', 'lightcoral', 'lightskyblue'])
axes[1, 1].set_title('Prediction Confidence Distribution', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

# Save optimization results
results_df.to_csv('optimization_results.csv', index=False, encoding='utf-8')
print(f"\nðŸ’¾ Optimization results saved to: optimization_results.csv")
```
